```python
def scrape_to_dict_list(url, list_of_slugs):
    dict_list = []

    for slug in list_of_slugs:
        sleep(1)
        try:
            res = requests.get(url + slug)
            soup = BeautifulSoup(res.content)
            
            print(f'{slug}')

            tbody = soup.find('tbody')
        

        

            for row in tbody.find_all('tr'):
                inmates = {}
                inmates['name'] = row.find('td').text.strip()
                inmates['age'] = row.find('td', {'data-title': 'Age'}).text
                inmates['main_crime'] = row.find('td', {'data-title': 'Main Crime'}).text
                inmates['entered_on'] = row.find('td', {'data-title': 'Entered On'}).text
                inmates['term'] = row.find('td', {'data-title': 'Term'}).text
                inmates['crime_location'] = row.find('td', {'data-title': 'Crime Location'}).text
                inmates['home_county'] = row.find('td', {'data-title': 'Home County'}).text

                dict_list.append(inmates)
        
        except:
            print('There are no more pages. ')
            break



    print('Finished.')

    return dict_list
```

### Trying to take this challenge in steps. Ultimately want to do it in one function, but not there yet.

Step 1: Get the terms that have 'years' to a string of the number only. 

```python
#tried to do multiple items in the for loop
#for me, had to come back to pulling out years first, otherwise getting errors 
term_list = []
for term in df['term']:
    
    no_years = term.replace(' years', '')
    
    term_list.append(no_years)
```

```python
df['term_rd2'] = term_list
```

```python
df.term_rd2.unique()
```

```python
#getting the 6 months into .5s to be able to change to floats
#have to keep them ALL STRINGS until can CONVERT ALL AT ONCE 
term_list_2 = []

for term in df['term_rd2']:
    
    if ', 6 months' in term:
        
        no_months_main = term.replace(', 6 months', '.5')
        
        term_list_2.append(no_months_main)
    elif ' 7 months' in term:
        #need to get them to strings that can be converted
        no_months = term.replace(', 7 months', '.583')
        
        term_list_2.append(no_months)
   
    
    else:

        term_list_2.append(term)
```

```python
df['term_rd2'] = term_list_2
```

```python
df.term_rd2.unique()
```

```python
#trying to do the remaining months and days together, if possible
list_3 = []
for term in df['term_rd2']:
    if '7 months' in term:
        no_months = term.replace('7 months', '.583')
        
        list_3.append(no_months)
    
    
    elif ' year' in term:
        no_year = term.replace(' year', '')
        
        list_3.append(no_year)
    
    elif '3 months' in term:
        no_days = term.replace(', 3 months, 25 days', '.318')
        
        list_3.append(no_days)
        
    else:
        
        list_3.append(term)
```

```python
df['term_rd2'] = list_3
```

```python
df.term_rd2.unique()
```

```python
#saving the word conversion for last
new_list = []
for term in df['term_rd2']:
    if term == 'Life Without Parole':
        number = term.replace('Life Without Parole', '500')
    
    
        new_list.append(number)
    elif term == 'Life':
        
        number_2 = term.replace('Life', '150')
        
        new_list.append(number_2)
    else:
        
        new_list.append(term)
```

```python
df['term_rd2'] = new_list
```

```python
df.term_rd2.unique()
```

```python
#this is the last step, once they have all been replaced and are float-ready 
floats_list = []

for term in df.term_rd2:
    one = float(term)
        
    floats_list.append(one)
    
 
         
```

```python
df['term_rd2'] = floats_list
```
